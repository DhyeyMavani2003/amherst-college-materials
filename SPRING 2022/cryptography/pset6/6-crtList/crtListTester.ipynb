For example, $\texttt{crtList( [(2,3), (3,5), (0,2)] )}$ should return $(8, 30)$, since the system of three congruences $x \equiv 2 \pmod{3}, x \equiv 3 \pmod{5}, x \equiv 0 \pmod{2}$ is equivalent to the single congruence $x \equiv 8 \pmod{30}$. 

The integer $a$ should be reduced modulo $m$, i.e. $0 \leq a < m$. The moduli $m_i$ will be integers up to $256$ bits in length, and the list will contain up to $128$ entries.

```python
import numpy
def num(a,b):
    tot = 0
    while (b>=a):
        tot += (1/numpy.log(a))
        a += 1
    return tot
```

```python
def isWitness(a,n):
    # Divide (n-1) by 2 until it's odd
    k = 0
    q = n - 1
    while q % 2 == 0:
        q = q//2
        k+=1
    b = pow(a,q,n)
    if (b == 1):
        return False
    for u in range(k):
        if b == n-1:
            return False
        b = b*b % n
    return True

def countWitnesses(n):
    count = 0
    for a in range(1,n):
        if isWitness(a,n):
            count += 1
    return count

for x in range(1000000,1000010):
    print(countWitnesses(x)/x)
    
```
