# Exercise 1
This exercise reconciles two points of view on consecutive trades:

1. One can treat the second order as the first order's continuation.
2. One can treat the second order as a separate trade from an external source.

One shows that the two points of view lead to the same optimal trading strategy. Consider a generalized OW model with parameters $\beta_t >0$, $\lambda_t = e^{\gamma_t}$ satisfying the no price-manipulation condition $2\beta_t + \gamma_t'>0$. For a given trading process $Q$, the price impact is the solution to the SDE

$$
dI_t = -\beta_t I_t dt + \lambda_{t-} dQ_t.
$$

Consider a deterministic alpha signal $\alpha_t$ (with $\alpha_T = 0$).

## Exercise 1.1
In the continuation approach, $Q_{0-} = \tilde{Q}$ and $I_{0-} =\tilde{I}$ are non-zero and capture the first order's effect. Derive the optimal impact state $I^0_t$.


<span style="color:brown"> 
Note that we use the left limits to denote quantities before the block trades: for example, $Q_{0-}$ is the position before the block trade at $t=0$, and $Q_0$ is the position after the block trade. When there is no block trade at $t = 0$, we have $Q_{0-} = Q_0$.
</span>

## Solution:
The following is a derivation similar to Week 4 Lecture Notes, pg. 25-31. We aim to derive the formula of optimal impacts on Week 7 Notes, pg. 19.

We want to maximize the wealth equation over $Q = (Q_t)_{0 \leq t \leq T}$ (in particular, $Q_0$ determines the size of the block trade at $t = 0$):

$$
    \int_0^T (\alpha_{t-} - I_{t-}) \, dQ_t - \frac{1}{2}[I, Q]_T + [\alpha, Q]_T.
$$

After applying the transformation

$$
    dQ_t = \frac{1}{\lambda_{t-}} \left( dI_t + \beta_t I_t dt \right) = e^{-\gamma_{t-}} \left( dI_t + \beta_t I_t dt \right),
$$

the wealth equation is

$$
     \int_0^T e^{-\gamma_{t-}} (\alpha_{t-} - I_{t-}) \left( dI_t + \beta_t I_t dt \right) - \frac{1}{2} \int_0^T e^{-\gamma_t} \, d[I, I]_t + \int_0^T e^{-\gamma_t} \, d[\alpha, I]_t.
$$

Note that when integrating against a semi-martingale (e.g., $Q$ and $I$), the integrand needs to be the left limits (e.g. $I_{t-}$); however, when integrating against $dt$ or a process of finite variation (e.g., $d[I, I]_t$), it does not change the result whether one uses $I_t$ or $I_{t-}$.

Using integration by parts from $t = 0-$ to $t=T$ (e.g., applying Ito's lemma to $e^{-\gamma_t} a_t I_t$ and $\frac{1}{2}e^{-\gamma t} I_t^2$ and writing in the integral form), we have 

$$
    e^{-\gamma_T} a_T I_T - e^{-\gamma_{0-}} a_{0-} I_{0-} = -\int_0^T \gamma_t' e^{-\gamma_{t}} \alpha_t I_t \, dt + \int_0^T e^{-\gamma_{t-}} \alpha_{t-} dI_t + \int_0^T e^{-\gamma_{t-}} I_{t-} d \alpha_t +  \int_0^T e^{-\gamma_{t}}d[\alpha, I]_t
$$

and

$$
    \frac{1}{2}(e^{-\gamma_{T}} I_T^2 - e^{-\gamma_{0-}} I_{0-}^2) = -\frac{1}{2} \int_0^T \gamma_t' e^{-\gamma_{t}} I_t^2 \, dt + \int_0^T e^{-\gamma_{t-}} I_{t-} dI_t + \frac{1}{2} \int_0^T e^{-\gamma_{t}} \, d[I, I]_t.
$$

Furthermore, we assume that $a_t$ is deterministic and differentiable in time. Then, we have $d \alpha_t = \alpha_t' dt.$

Then, the wealth equation can be turned into

$$
     e^{-\gamma_T} a_T I_T - e^{-\gamma_{0-}} a_{0-} I_{0-} + \int_0^T e^{-\gamma_{t}} I_t (\gamma_t' \alpha_t - \alpha_t') \, dt +  \int_0^T \beta_t e^{-\gamma_{t}} (\alpha_t  I_t - I_t^2)\, dt - \frac{1}{2} \left( e^{-\gamma_{T}} I_T^2 - e^{-\gamma_{0-}} I_{0-}^2 + \int_0^T \gamma_t' e^{-\gamma_{t}} I_t^2 \, dt \right).
$$

 Now, if we optimize the expression over $I = (I_t)_{0\leq t \leq T}$, the problem is myopic:

$$
    \forall t \in [0, T), \quad I_t^* = \frac{(\beta_t + \gamma_t')\alpha_t - \alpha_t'}{2\beta_t + \gamma_t'}; \quad I_T^* = \alpha_T.
$$

## Solution 1.1:

<span style="color:blue">
Thus, the optimal impact state with initial impact $I_{0-} = \tilde{I}$ and
    
$$
    \forall t \in [0, T), \quad I_t^0 = \frac{(\beta_t + \gamma_t')\alpha_t - \alpha_t'}{2\beta_t + \gamma_t'}; \quad I_T^0 = \alpha_T.
$$
    
In particular, the jump of the impact state at $t = 0$ is $I_0^0 - \tilde{I}$.
</span>

## Exercise 1.2
In the separate approach, $Q_0 = 0$ and $I_0 =0$. The external impact $\bar{I}$ captures the first order's effect. The first order's external impact satisfies

$$
\quad d\bar{I}_t = -\beta_t \bar{I}_t dt, \quad t \geq 0-
$$
with
$ \bar{I}_{0-} = \tilde{I}$. Derive the optimal impact state $I^1_t$.

## Solution 1.2:
<span style="color:blue">
We repeat the derivation in Exercise 1.1, replacing $\alpha_t$ by $\alpha_t - I_t$ and $\alpha_t'$ by $\alpha_t' - \bar{I}_t'$ verbatim. The interpretation is that while we capture $\alpha_t$, we need to pay an external (uncontrolled) impact $\bar{I}_t$ in addition to an internal (controlled) impact $I_t$. 
    
The optimal impact state is
    \begin{equation*}
    \forall t \in [0, T), \quad I^1_t = \frac{\beta_t + \gamma'_t}{2\beta_t + \gamma'_t}\left(\alpha_t - \bar{I}_t\right) - \frac{1}{2\beta_t + \gamma'_t}\left(\alpha'_t -\bar{I}'_t\right) ; \quad I^1_T = \alpha_T - \bar{I}_T.
    \end{equation*}
    
Note that $\bar{I}_t' = - \beta_t \bar{I}_t$, so
$$
    \frac{\beta_t + \gamma'_t}{2\beta_t + \gamma'_t} \bar{I}_t - \frac{1}{2\beta_t + \gamma'_t} \bar{I}'_t = \bar{I}_t,
$$
and it follows that 
    
$$I^1_t = I^0_t - \bar{I}_t.$$
    
In particular, the jump of impact $I^1$ at $t = 0$ is the same as that of $I^0$, since $\bar{I}_{0-} = \bar{I}_0$.

</span>

## Exercise 1.3
Reconcile the two points of view.

## Solution 1.3:

<span style="color:blue">

The equation $I^1 = I^0 - \bar{I}$ yields two implications:

1. The total impact state is equal in both cases. The chosen point of view shifts the impact of the first order from internal to external.

    
2. The corresponding optimal trading processes $Q^0$ and $Q^1$ are identical. Indeed,
    
\begin{align*}
\lambda_t dQ^0_t &= \beta_t I^0_t dt + dI^0_t \\
							 &= \beta_t (I^1_t + \bar{I}) dt + dI^1_t + d\bar{I}_t \\
							 &= \beta_t I^1_t dt + dI^1_t \\
							 &= \lambda_t dQ^1_t
\end{align*}
    
for $t \in [0,T)$. Manual inspection of the jumps at $t= 0$ and $t=T$ confirms that they match across both points of view.    
</span>

## Exercise 2
This exercise finds a lower bound on the decay parameter $\beta_t$ of a generalized OW model based on a liquidity profile $v_t$ to rule out price manipulation strategies. Assume the intraday volume profile $v$ follows the deterministic curve

\begin{equation*}
v_t = e^{4\cdot (t- 0.7)^2}.
\end{equation*}

*Local square root model under the calendar clock:* 

The impact model considered is
$$
dI_t = -\beta I_t dt + \frac{\lambda}{\sqrt{v_t}} dQ_t.
$$

1.1 Establish the lower bound on $\beta$ that guarantees no price manipulation.

1.2 For $\beta = 0.01$, find a pair of trades that lead to a price manipulation paradox.


*Local log model under the volume clock:*

The impact model considered is 
$$
dI_t = -\beta v_t I_t dt + \frac{\lambda}{v_t} dQ_t.
$$

2.1. Establish a lower bound on $\beta$ that guarantees no price manipulation.

2.2. For $\beta = 0.01$, find a pair of trades that lead to a price manipulation paradox.

## Solution 2.1.1:
<span style="color:blue">
In this model, we note that $\beta_t = \beta$ and $\lambda_t = \lambda/\sqrt{v_t}$, so $\gamma_t = \log \lambda - \frac{1}{2} \log v_t$. 
Plugging in the formula of $v_t$, we have
    
\begin{equation*}
\forall t \in [0,1] \quad \gamma'_t = -4(t - 0.7) \geq - 1.2.
\end{equation*}
    
Therefore, The no price manipulation condition $2\beta_t + \gamma_t' > 0$ yield that the lower bound on $\beta$
    
\begin{align*}
   \beta > \max_{t \in [0,1]}(-0.5 \gamma_t') = \max_{t \in [0,1]} 2(t-0.7) = 0.6.
\end{align*}
    
This lower bound corresponds to a maximum half-life for the price impact model of around $0.73$ days. Therefore, a price impact model that decays slower than this time scale exhibits price manipulation.
</span>

## Solution 2.1.2:
<span style="color:blue">
One wishes to enter and exit a position during a period where $\gamma'$ is strongly negative. For example, $\gamma'$ is negative over the interval $[0.7, 1]$. Consider the following trade pair:
    
1. Buy one share at time $t = 0.7$.
    
2. Sell one share at time $t = 1$.
 
Recall that the transaction costs in general are
    
$$
\int_0^T I_{t-} dQ_t + \frac{1}{2}[I, Q]_T. 
$$
    
For discretized trades, $[I, Q]_T$ is $\sum_{t} \Delta_t I \, \Delta_t Q$, where $\Delta_t Q = Q_t - Q_{t-}$ and $\Delta_t I = I_t - I_{t-}$.
 
The corresponding transaction costs are
\begin{align*}
I_{0.7-} \Delta_{0.7} Q &+ I_{1-} \Delta_{1} Q + \frac{1}{2}\Delta_{0.7} I\Delta_{0.7} Q + \frac{1}{2}\Delta_{1} I\Delta_{1} Q \\
\quad &= (0) - \lambda e^{-0.3\cdot 0.01} + \frac{\lambda}{2} + \frac{\lambda}{2} e^{-2(0.3)^2}\\
&= \frac{\lambda}{2}(1 + e^{-0.18} - 2 e^{-0.003})\\
&<0
\end{align*}
as $1 + e^{-0.18} - 2 e^{-0.003} \approx -0.15$.    
</span>

## Solution 2.2.1:
<span style="color:blue">
In this model, we note that $\beta_t = \beta v_t$ and $\lambda_t = \lambda/v_t$, so $\gamma_t = \log \lambda - \log v_t$. 
Plugging in the formula of $v_t$, we have
    
\begin{equation*}
\forall t \in [0,1] \quad \gamma'_t = -8(t - 0.7) \geq - 2.4.
\end{equation*}
    
Therefore, The no price manipulation condition $2\beta_t + \gamma_t' > 0$ yield that the lower bound on $\beta$
    
\begin{align*}
   \beta > \max_{t \in [0,1]}(-0.5 \gamma_t' v_t^{-1}) = \max_{t \in [0,1]} 4(t-0.7)e^{-4(t-0.7)^2} \approx 0.8372.
\end{align*}
</span>

## Solution 2.2.2:
<span style="color:blue">
Consider the following trade pair:
    
1. Buy one share at time $t = 0.7$.
    
2. Sell one share at time $t = 1$.
 
For $t \in (0.7, 1)$, 
\begin{align*}
dI_t &= -\beta e^{4\cdot(t-0.7)^2} I_t dt \\
     &> -\beta e^{0.36} I_t dt \\
		 &> -2 \beta I_t dt 
\end{align*}
and $I_{0.7} = \lambda$.
Hence,
    
\begin{equation*}
I_1 > \lambda e^{2\cdot 0.01\cdot 0.3}
\end{equation*}
    
The corresponding transaction costs are
\begin{align*}
I_{0.7-} \Delta_{0.7} Q &+ I_{1-} \Delta_{1} Q + \frac{1}{2}\Delta_{0.7} I\Delta_{0.7} Q + \frac{1}{2}\Delta_{1} I\Delta_{1} Q \\
\quad &= (0) - I_1 + \frac{\lambda}{2} + \frac{\lambda}{2} e^{-4(0.3)^2}\\
&< \frac{\lambda}{2}(1 + e^{-0.36} - 2 e^{-0.006})\\
&<0
\end{align*}
as $1 + e^{-0.36} - 2 e^{-0.006} \approx -0.29$.
</span>

## Exercise 3:

This exercise solves an optimal execution problem under a globally concave AFS price impact model. The AFS model is particularly relevant when submitting sizable orders. In that regime, the instantaneous liquidity conditions are of second order, and price impact's concavity drives trading costs. The interval $[0,T]$ represents a single trading day. Consider
\begin{equation*}
I_t = h(J_t)
\end{equation*}
with $h(x) = \text{sign}(x)|x|^c$ and local dynamics
\begin{equation*}
dJ_t = - \beta J_t dt + \lambda dQ_t 
\end{equation*}
for $c \in (0,1]$. The corresponding self-financing equation is
\begin{equation*}
dY_t = Q_t dS_t - h({J}_t) dQ_t - \frac{\lambda}{2} h'({J}_t) d[Q,Q]_t.
\end{equation*}

Assume that $I_0 = 0$.

1. Under this model, solve the optimal execution problem for the target impact state $I^*.$ Assume given a deterministic alpha $\alpha_t$ and an overnight alpha $\tilde{\alpha}$.

2. What is the relationship between the order size $Q_T$ and the implied overnight alpha $\tilde{\alpha}$ when there is no trading alpha?

3. Assume that $I_0>0$ from a previous order. Solve the idealized optimal execution problem for the target impact state $I^*$ without trading alpha. How does the order size $Q_T$ change given $I_0$?

## Solution 3.1
<span style="color:blue">
Recall that in Exercise 2 of Week 4 Assignment, the wealth equation is
$$
E\left[Y_T\right] = E\left[\int_0^T  \left(\beta \alpha_t J_t - \alpha'_t J_t- \beta h(J_t)J_t\right) dt + \alpha_T J_T -  \frac{1}{1+c}|J_T|^{1+c} \right].
$$

Now, in the presence of overnight alpha, we have an additional term in the wealth equation
$$
    \tilde \alpha Q_T = \frac{1}{\lambda} \left[ \int_0^T \beta \tilde{\alpha} J_t dt + \tilde{\alpha} J_T \right].
$$

Thus, the objective function to maximize is 
$$
E\left[Y_T\right] = \frac{1}{\lambda}E\left[\int_0^T \left(\left(\tilde{\alpha} + \alpha_t\right)\beta J_t - \alpha'_t J_t - \beta \left|J_t\right|^{1+c}\right) dt + (\tilde{\alpha} + \alpha_T) J_T - \frac{1}{1+c}\left|J_T\right|^{1+c}\right].
$$

The problem is myopic in $I$, and one solves it pointwise:
\begin{equation*}
\forall t\in (0,T), \quad I^*_t = \frac{1}{1+c}\left(\tilde{\alpha} + \alpha_t - \beta^{-1} \alpha'_t\right); \quad I^*_T = \tilde{\alpha} + \alpha_T.
\end{equation*}
As a sanity check, one recovers the original OW model's solution for $c = 1$. 
</span>

## Solution 3.2
<span style="color:blue">
The implied execution problem is a special case where the intraday alpha $\alpha$ is absent. The target impact state is
\begin{equation*}
\forall t\in (0,T), \quad I^*_t = \frac{1}{1+c}\tilde{\alpha}; \quad I^*_T = \tilde{\alpha}.
\end{equation*}

This relationship suggests that the ratio between the target impact state and the overnight or implied alpha should be $\frac{1}{1+c}$. For $c <1$, this ratio is higher than in the original OW model.

One relates the overnight alpha $\tilde{\alpha}$ to the terminal position $Q_T$. Assume that $\tilde{\alpha}>0$ for simplicity. Then, the order size satisfies
\begin{align*}
Q_T &= \frac{1}{\lambda}J_T + \frac{\beta}{\lambda}\int_0^T J_t dt \\
		&= \frac{1}{\lambda}\tilde{\alpha}^{1/c} + \frac{\beta T}{\lambda}\left(\frac{\tilde{\alpha}}{1+c}\right)^{1/c} \\
	  &= \left(1 + \frac{\beta T}{(1+c)^{1/c}}\right)	\frac{\tilde{\alpha}^{1/c}}{\lambda}.
\end{align*}
Equivalently, one defines the implied alpha $\tilde{\alpha}$ of the target position $Q_T$ as 
\begin{equation*}
\tilde{\alpha} = (1+c) \left(\frac{\lambda Q_T}{(1+c)^{1/c} + \beta T}\right)^c.
\end{equation*}
</span>

## Solution 3.3
<span style="color:blue">
By the myopic nature of the control problem in $I$, the target impact state remains unchanged:
\begin{equation*}
\forall t\in (0,T), \quad I^*_t = \frac{1}{1+c}\tilde{\alpha}; \quad I^*_T = \tilde{\alpha}.
\end{equation*}

Therefore, the optimal strategy in trade space is identical for $t \in (0,T]$, and only differs by a jump at time $t = 0$. Integrating 
    
$$
dJ_t = -\beta J_t dt + \lambda dQ_t
$$
from $0-$ to $T$ yields
\begin{equation*}
J_T - J_{0-} + \beta \int_0^T J_t dt = \lambda Q_T.
\end{equation*}
Hence, the order size $Q_T$ decreases by $\frac{1}{\lambda}J_{0-}$ to
$$
\frac{1}{\lambda} \left[ \left(1 + \frac{\beta T}{(1+c)^{1/c}}\right) \tilde{\alpha}^{1/c} - I_{0-}^{1/c} \right],
$$    
and the modified implied alpha relationship is
\begin{equation*}
\tilde{\alpha} = (1+c)\left(\frac{\lambda Q + I^{1/c}_0}{(1+c)^{1/c} + \beta T}\right)^c.
\end{equation*}
</span>
