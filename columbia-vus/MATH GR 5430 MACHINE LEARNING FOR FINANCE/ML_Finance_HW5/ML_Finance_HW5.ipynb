# 1. Setup and Data Fetching (5)

#### • Obtain and import 1 stock dataset of your choice from yfinance as a Pandas dataframe.

```python
import yfinance as yf
import pandas as pd

# Define the ticker symbol
tickerSymbol = 'AAPL'

# Get data on this ticker
tickerData = yf.Ticker(tickerSymbol)

# Get the historical prices for this ticker
tickerDf = tickerData.history(period='1d', start='2010-1-1', end='2023-5-5')

# Print the dataframe
tickerDf
```

```python
# Fill missing values using forward fill
tickerDf.fillna(method='ffill', inplace=True)
tickerDf
```

#### • Plot the time series of adjusted closing stock prices. Plot the time series of stock returns. What do you discover about those variables?

```python
import matplotlib.pyplot as plt

# Plot the time series of adjusted closing stock prices
plt.figure(figsize=(10, 5))
plt.plot(tickerDf['Close'])
plt.title('Adjusted Closing Stock Prices')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()

# Calculate daily returns and plot the time series of stock returns
daily_returns = tickerDf['Close'].pct_change()
plt.figure(figsize=(10, 5))
plt.plot(daily_returns)
plt.title('Daily Returns')
plt.xlabel('Date')
plt.ylabel('Return')
plt.show()
```

We can see a seasonal variation of the returns and it is evident that the volatility spiked in tbe more recent years compared to previous years!

#### • Is the stock price stationary? Is the stock return stationary? If not, transform the return data such that it is a stationary time series.

```python
from statsmodels.tsa.stattools import adfuller

# Perform ADF test on the stock price series
result_price = adfuller(tickerDf['Close'])
print('ADF Statistic for stock price: {:.5f}'.format(result_price[0]))
print('p-value for stock price: {:.5f}'.format(result_price[1]))
print('Critical Values for stock price:')
for key, value in result_price[4].items():
    print('\t{}: {}'.format(key, value))

# Calculate daily returns and perform ADF test on the returns series
daily_returns = tickerDf['Close'].pct_change().dropna()
result_returns = adfuller(daily_returns)
print('ADF Statistic for stock returns: {:.5f}'.format(result_returns[0]))
print('p-value for stock returns: {:.5f}'.format(result_returns[1]))
print('Critical Values for stock returns:')
for key, value in result_returns[4].items():
    print('\t{}: {}'.format(key, value))
```

We can see that the stock price is definitely not stationary, but the stock returns are stationary with over 99% confidence because the p-value is even less than 0.00001

# 3. Time Series Model Configuration and Testing (50)

#### • Draw the ACF and PACF graphs of the return time series to determine p and q

```python
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Calculate daily returns
daily_returns = tickerDf['Close'].pct_change().dropna()

# Plot the ACF and PACF graphs
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(10,8))
plot_acf(daily_returns, ax=ax[0])
plot_pacf(daily_returns, ax=ax[1])
plt.show()
```

The resulting ACF and PACF graphs can help us identify the appropriate values of p and q for an ARIMA model. Specifically, the PACF plot can help identify the appropriate value of p, while the ACF plot can help identify the appropriate value of q.

If the PACF plot shows a significant spike at lag k and the ACF plot shows a gradual decay, this indicates that an AR(p) model may be appropriate, where p is the lag value corresponding to the significant spike in the PACF plot.

If the ACF plot shows a significant spike at lag k and the PACF plot shows a gradual decay, this indicates that an MA(q) model may be appropriate, where q is the lag value corresponding to the significant spike in the ACF plot.

However, in practice, the selection of p and q values can be more complex, and often requires iteratively fitting and evaluating multiple ARIMA models with different parameter combinations.

#### • Visualize first order and second order differencing using diff() to determine d

```python
# Compute first order difference
diff1 = tickerDf['Close'].diff().dropna()

# Compute second order difference
diff2 = tickerDf['Close'].diff().diff().dropna()

# Plot the time series and the first and second order differences
fig, ax = plt.subplots(nrows=3, ncols=1, figsize=(10,8))
tickerDf['Close'].plot(ax=ax[0])
ax[0].set_title('AAPL Close Price')
diff1.plot(ax=ax[1])
ax[1].set_title('First Order Difference')
diff2.plot(ax=ax[2])
ax[2].set_title('Second Order Difference')
plt.tight_layout()
plt.show()
```

If the first order difference plot shows a constant mean and variance, then d=1 is appropriate. If the second order difference plot shows a constant mean and variance, then d=2 is appropriate. If neither of the plots show a constant mean and variance, then further differencing may be required.

#### • Fit an ARIMA(p,d,q) model (Hint: from statsmodels.tsa.arima.model import ARIMA) How does the model perform? Print a summary of our fitting. Are the coefficients statistically significant?

```python
import statsmodels.api as sm
import warnings

# Ignore all warnings
warnings.filterwarnings('ignore')

# Fit the ARIMA model
model = sm.tsa.ARIMA(tickerDf['Close'], order=(1,1,1))
result = model.fit()

# Print the summary of the model
print(result.summary())
```

We can see that the SARIMAX performs very well and all the coefficients are significant with p-values even less than 0.001

#### • Plot the residual time series and the density plot of the residuals. What do you discover?

```python
# Plot the residual time series
residuals = result.resid
fig, ax = plt.subplots(figsize=(10, 5))
residuals.plot(ax=ax)
ax.set_title('Residuals of ARIMA Model')
ax.set_xlabel('Date')
ax.set_ylabel('Residuals')
plt.show()

# Plot the density plot of the residuals
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(111)
sm.graphics.tsa.plot_acf(residuals, lags=40, ax=ax)
ax.set_title('Autocorrelation Plot of Residuals')
plt.show()

# Plot the density plot of the residuals
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(111)
sm.graphics.tsa.plot_pacf(residuals, lags=40, ax=ax)
ax.set_title('Partial Autocorrelation Plot of Residuals')
plt.show()
```

The residual time series plot shows the residuals over time, which should appear as white noise with mean zero and constant variance if the model is a good fit. The density plot of the residuals should show a normal distribution if the residuals are normally distributed.

By looking at the residual time series and density plot, we can determine if the ARIMA model is a good fit for the data. If the residual time series exhibits a pattern or the density plot is not close to normal, this suggests that the model may need to be revised.

#### • Fit another model without the last 10 timestamps and use your model to rolling forecast 10 timestamps forward. How well does your model predict? Plot the result and calculate the RMSE

```python
import yfinance as yf
import statsmodels.api as sm
import numpy as np
import matplotlib.pyplot as plt

# Define the ticker symbol
tickerSymbol = 'AAPL'

# Get data on this ticker
tickerData = yf.Ticker(tickerSymbol)

# Get the historical prices for this ticker
tickerDf = tickerData.history(period='1d', start='2010-1-1', end='2023-5-5')

# Fit the ARIMA model with the last 10 timestamps removed
train_data = tickerDf.iloc[:-10, :]
model = sm.tsa.ARIMA(train_data['Close'], order=(1,1,1))
result = model.fit()

# Use the model to rolling forecast 10 timestamps forward
forecast_start = len(train_data)
forecast_end = len(tickerDf) - 1
forecast = result.forecast(steps=10)
print(forecast)
# Plot the actual and forecasted values
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(tickerDf.index[forecast_start:],
        np.array(forecast),
        label='Forecast')
ax.plot(tickerDf.index[forecast_start:], tickerDf['Close'][forecast_start:], label='Actual')
ax.set_title('ARIMA Model Forecast')
ax.set_xlabel('Date')
ax.set_ylabel('Price')
ax.legend()
plt.show()

# Calculate the RMSE
mse = ((np.array(forecast) - tickerDf['Close'][forecast_start:])**2).mean()
rmse = np.sqrt(mse)
print('RMSE:', rmse)
```

# 4. RL Basics (20)

#### • What is reinforcement learning? How does it differ from supervised learning?

Reinforcement learning is a type of machine learning where an agent learns to interact with an environment by taking actions and receiving rewards or punishments based on those actions. The goal of reinforcement learning is to find a policy or set of actions that maximizes a cumulative reward over time.

In contrast, supervised learning is a type of machine learning where an algorithm learns from a labeled dataset to predict an output value given an input. The goal of supervised learning is to generalize from the training data to make accurate predictions on new, unseen data.

The main difference between reinforcement learning and supervised learning is that in reinforcement learning, the agent learns from its own experiences in the environment through trial and error, while in supervised learning, the algorithm learns from labeled examples provided by a human expert.

Another difference is that in reinforcement learning, there is no explicit correct answer or target variable for the agent to learn, whereas in supervised learning, the correct answer is provided in the labeled data. Instead, in reinforcement learning, the agent learns to maximize a reward signal that is provided by the environment, which may be delayed or stochastic.

Overall, reinforcement learning is used in applications where the optimal actions are not known in advance and must be learned through interaction with the environment, while supervised learning is used in applications where labeled data is available and the goal is to make accurate predictions on new data.

#### • Formally define an RL problem

Reinforcement learning can be formalized as a sequential decision-making problem. It consists of an agent that interacts with an environment over a sequence of discrete time steps. At each time step, the agent observes the current state of the environment and selects an action to perform based on the current state. The environment then transitions to a new state and returns a reward signal to the agent, which may depend on the current state and the action taken.

The goal of the agent is to learn a policy, which maps states to actions, that maximizes the cumulative reward over time. The policy is usually represented by a function that takes the current state as input and outputs a probability distribution over possible actions. The agent selects an action according to the policy and the environment transitions to a new state. The agent updates its policy based on the observed rewards and the new state, using a learning algorithm such as Q-learning, SARSA, or policy gradients.

The RL problem can be formalized as a Markov decision process (MDP), which is a tuple (S, A, P, R, γ), where S is the set of possible states, A is the set of possible actions, P is the state transition probability function, R is the reward function, and γ is the discount factor. The state transition function $P(s'|s,a)$ specifies the probability of transitioning to state s' given the current state s and action a. The reward function R(s,a,s') specifies the reward received for transitioning from state s to state s' by taking action a. The discount factor γ determines the importance of future rewards relative to immediate rewards. The objective of the RL agent is to learn a policy π that maximizes the expected cumulative reward:

$$E[∑_{t=0}^∞ γ^t r_t]$$,

where $r_t$ is the reward received at time step t.

In summary, the RL problem is a sequential decision-making problem where an agent learns to interact with an environment to maximize a cumulative reward over time. It is formalized as a Markov decision process, where the agent learns a policy that maps states to actions based on the observed rewards and state transitions.

#### • What is the Markov property? How is it useful in RL?

The Markov property is a property of stochastic processes that states that the future state of the process depends only on the current state and not on the past states. In other words, the current state of the process encapsulates all the relevant information needed to predict the future behavior of the process.

In the context of RL, the Markov property is particularly useful because it allows the agent to make decisions based only on the current state of the environment, without needing to consider the entire history of past states and actions. This makes the problem of learning a policy more tractable, since the agent can focus on learning to select actions that are optimal based on the current state of the environment.

Formally, the Markov property is formalized as the Markov assumption in the definition of a Markov decision process (MDP), which is a key framework used in RL. The Markov assumption states that the state transition probability function P(s'|s,a) depends only on the current state s and action a, and not on any past states or actions. This allows the agent to use the current state as a sufficient statistic to make predictions about the future behavior of the environment, and to learn a policy that maps states to actions based only on the current state.

In summary, the Markov property is a fundamental concept in stochastic processes, and it is particularly useful in RL because it allows the agent to make decisions based only on the current state of the environment, without needing to consider the entire history of past states and actions. This simplifies the problem of learning a policy and makes it more tractable.

#### • In simple terms, what is Q-learning?

Q-learning is a model-free, off-policy RL algorithm used to learn the optimal action-value function Q(s,a) that maps states and actions to expected future rewards. The goal of Q-learning is to learn a policy that maximizes the total expected reward obtained by the agent over time.

The Q-learning algorithm works by iteratively updating the estimate of the optimal action-value function using the Bellman equation. At each time step, the agent observes the current state of the environment, selects an action based on an exploration-exploitation tradeoff, receives a reward, and transitions to the next state. The Q-value of the current state-action pair is updated using the reward and the estimate of the maximum Q-value of the next state. The update equation is:

$$Q(s,a) ← Q(s,a) + α [r + γ maxa' Q(s',a') - Q(s,a)]$$

where Q(s,a) is the current estimate of the Q-value for state s and action a, r is the reward obtained by taking action a in state s, s' is the next state, a' is the next action, γ is a discount factor that controls the tradeoff between immediate and future rewards, and α is a learning rate that controls the step size of the update.

Over time, the Q-values are updated based on the agent's experience of interacting with the environment, and the learned Q-values are used to select the optimal policy that maximizes the expected reward. Q-learning is a powerful and widely used RL algorithm that has been successfully applied to a wide range of tasks in robotics, game playing, and other domains.
