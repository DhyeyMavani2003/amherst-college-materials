# Question 1

This question is to investigate and compare the 10-day VaR calculations using a 10-day window and a 1-day
window. We will use the Dow Jones Index closing prices from Yahoo Finance as the data source.

Hypothetically, suppose we have held a portfolio of 1 share of DJI for the past 2 years.

The current date is April 21st 2023.

```python
import yfinance as yf
import pandas as pd
import numpy as np

start_date = "2021-04-21"
end_date = "2023-04-22"
dji = yf.download("^DJI", start = start_date, end = end_date)["Adj Close"]
dji
```

```python
dji_returns = np.log(dji) - np.log(dji.shift(1))
dji_returns
```

```python
dji_returns_ewm = dji_returns.ewm(alpha=0.06, adjust=True)
```

```python
dji_ewm_mean = dji_returns_ewm.mean()
dji_ewm_std = dji_returns_ewm.std()
```

```python
# 99% parametric VaR
variance_p = dji_ewm_std**2
2.33*np.sqrt(variance_p[-1])
```

a) Compute the 95% 1-day historical VaR of the portfolio for the past 100 days. Each historical VaR value (say,
on date T) shall be computed using 250 days of daily log price returns (ie. returns on T-249, ..., T-1, T).
Display the result for the current date.

```python
dji_value = dji[-1]*np.exp(dji_returns)
dji_historical = dji_value - dji[-1]
dji_historical = dji_historical.dropna()
dji_historical
```

```python
dji_historical[-2:-1]
```

```python
np.percentile(np.array(dji_historical[-251:-1]),0.95)
```

```python
VaR_values = []
for i in range(1,101):
    dji_value = dji[-i]*np.exp(dji_returns)
    dji_historical = dji_value - dji[-i]
    dji_historical = dji_historical.dropna()
    VaR = np.percentile(np.array(dji_historical[-(i+250):-i]),0.95)
    VaR_values.append(VaR)
VaR_values
```

b) Estimate 95% 10-day VaR for the past 100 days from the results in (a). (Use: N-day VaR = 1-day VaR x ‚àöùëÅ)
Display the result for the current date.

```python
ten_day_VaR_values = VaR_values
for i in range(len(ten_day_VaR_values)):
    ten_day_VaR_values[i] *= 10
ten_day_VaR_values
```

```python
ten_day_VaR_values[0]
```

This question is about comparing different machine learning models in the prediction of option prices.
The attached file, option_data.csv, includes simulated prices of a European call option with different spot prices, strikes, time to expiry (in years) and volatilities.


a) Use option_data.csv as the training set to train the following models to predict option prices. Use grid
search with 5-fold cross validation to find the optimal hyperparameters when requested below.

i. SVM regression model: use RBF kernel with ùê∂ = 100, ùõæ = 0.5.

```python
######## SVM Model #####################
import sklearn
import numpy as np
import pandas as pd
import scipy.stats as stats
from sklearn import svm
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

import warnings
warnings.filterwarnings('ignore')

# y = option_price
# x = spot, strike, vol, time_to_expiry

# Load the option_data.csv file
option_data = pd.read_csv('option_data.csv')

# Split the data into training and testing sets
X = option_data.drop('option_price', axis=1)
y = option_data['option_price']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=8990)

from sklearn.preprocessing import StandardScaler 
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)

model_svm = svm.SVR(gamma=0.5, C=100) 
model_svm.fit(X_train, y_train)

y_predict = model_svm.predict(X_test[:5])
print(y_predict - y_test[:5])

print(f"Model score for the test data is {model_svm.score(X_test, y_test)}")

# cross validation
from sklearn.model_selection import cross_val_score 
cv_scores = cross_val_score(model_svm, X_train, y_train, cv=5)
print(f"Mean of the cross validation scores is {cv_scores.mean()}")
print(f"The cross validation scores are {cv_scores}")

# Grid search
from sklearn.model_selection import GridSearchCV 
from sklearn.model_selection import TimeSeriesSplit 

split = TimeSeriesSplit(n_splits=5)
parameter_grid = [{'C': [1.0, 10.0, 50.0, 100.0],
                   'gamma': [0.01, 0.1, 0.5, 1.0]}]
grid = GridSearchCV(model_svm,
                    param_grid = parameter_grid,
                    cv=5,
                    return_train_score=True)
grid.fit(X_train, y_train)
best_model_svm = grid.best_estimator_
print(f"Parameters of the best model from the GridSearch are {grid.best_params_}")
print(f"Best Score from the Grid Search is {grid.best_score_}")
##########################################
```

iii. Neural network with one hidden layer: use ReLU as activation function. Choose the best number ùëõ of hidden neurons from ùëõ = 8, 16, 32, 64, 128. For other parameters, feel free to pick your preferred ones.

```python
##### MLP Classifier Example from Class ###############
import sklearn
import numpy as np
from sklearn import datasets
from sklearn.datasets import make_classification

X, y = make_classification(n_samples=1000, n_features=2, n_classes=3, n_redundant=0, n_informative=2, n_clusters_per_class=1, random_state=0)

import matplotlib.pyplot as plt
colors = np.array(["blue", "orange", "pink"]) 
plt.scatter(X[:,0], X[:, 1], color = colors[y])

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

from sklearn.preprocessing import StandardScaler 
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)

print(X_train[:5])
print(y_train[:5])

from sklearn.neural_network import MLPClassifier 
mlp = MLPClassifier(hidden_layer_sizes=(8),
                    batch_size=32,#learning_rate= 'adaptive', # for 'sgd'
                    learning_rate_init=0.001, max_iter=1000, # no of epochs
                    early_stopping=True)

mlp.fit(X_train, y_train)

print(f"Neural Network MLP score is {mlp.score(X_test, y_test)}")

parm_grid = [{'hidden_layer_sizes': [(8), (16), (8,16,8)]}]
#######################################################################
```

```python
############ MLP Regressor Used Here ################
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler 
from sklearn.neural_network import MLPRegressor 
from sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings('ignore')

# Load data
option_data = pd.read_csv('option_data.csv')

# Split data into features (X) and target (y)
X = option_data.drop('option_price', axis=1)
y = option_data['option_price']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=8990)

# Scale the data
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)

# Define the MLPRegressor model with default hyperparameters
mlp = MLPRegressor(hidden_layer_sizes = [(8)],
                   batch_size = 32,
                   activation = "relu",
                   learning_rate_init = 0.01,
                   max_iter = 5000)

# Train the model on the training data
mlp.fit(X_train, y_train)

# Print the accuracy score of the model on the test data
print(f"Neural Network MLP score is {mlp.score(X_test, y_test)}")

# Define the grid of hyperparameters to search over
"""
parameter_grid = [{
    'hidden_layer_sizes': [(8), (16), (32), (64), (128)],
    'batch_size': [32, 64, 128],
    'learning_rate_init': [0.001, 0.01, 0.1],
    'max_iter': [1000, 5000, 10000]
}]
"""
parameter_grid = [{
    'hidden_layer_sizes': [(8), (16), (32), (64), (128)]
}]

# Perform grid search with 5-fold cross validation
grid_search = GridSearchCV(mlp,
                           param_grid = parameter_grid,
                           cv=5)
grid_search.fit(X_train, y_train)
best_model = grid_search.best_estimator_

# Print best parameters and score
print("Best parameters: ", grid_search.best_params_)
print("Best score: ", grid_search.best_score_)
#########################################################
```

```python
# Model 1
model_svm.fit(X, y)
test_pred_svm = model_svm.predict(test_data)

# Model 2
model_rf.fit(X, y)
test_pred_rf = model_rf.predict(test_data)

# Model 3
mlp.fit(X, y)
test_pred_mlp = mlp.predict(test_data)
```

```python
test_true = np.maximum(test_spot_price - test_strike, 0)
```
