```python
# parametric VaR calculation separately using matrix formula
import numpy as np
import pandas as pd

# Define the portfolio and historical dataframes
portfolio = {"JNJ": 1.2*10**6,
             "MRK": 2.5*10**6,
             "PFE": 3.5*10**6
            }
JNJ = pd.read_csv("hw3_JNJ.csv")
MRK = pd.read_csv("hw3_MRK.csv")
PFE = pd.read_csv("hw3_PFE.csv")

# Use the most recent market price as the current market price
current_prices = {"JNJ": JNJ.iloc[-1]["Adj Close"],
                  "MRK": MRK.iloc[-1]["Adj Close"],
                  "PFE": PFE.iloc[-1]["Adj Close"]
                 }

JNJ_r = np.log(JNJ["Adj Close"]) - np.log(JNJ["Adj Close"].shift(1))
JNJ_r.dropna()
JNJ_r_ewm = JNJ_r.ewm(alpha=0.06, adjust=True).mean().dropna()

MRK_r = np.log(MRK["Adj Close"]) - np.log(MRK["Adj Close"].shift(1))
MRK_r.dropna()
MRK_r_ewm = MRK_r.ewm(alpha=0.06, adjust=True).mean().dropna()

PFE_r = np.log(PFE["Adj Close"]) - np.log(PFE["Adj Close"].shift(1))
PFE_r.dropna()
PFE_r_ewm = PFE_r.ewm(alpha=0.06, adjust=True).mean().dropna()

# Use the exponentially weighted average of daily returns to estimate the variance-covariance matrix
returns_ewm = pd.concat([JNJ_r_ewm, MRK_r_ewm, PFE_r_ewm], axis=1)
cov_ewm = returns_ewm.cov()

# Calculate the portfolio variance and volatility
portfolio_var = np.dot(np.dot(list(portfolio.values()), cov_ewm), list(portfolio.values()))
portfolio_vol = np.sqrt(portfolio_var)

# Compute the 1-day VaR (with 99% confidence) for the portfolio
confidence_level = 0.99
z_score = 2.33  # Z-score for 99% confidence level
one_day_VaR = portfolio_vol * z_score

# Print the results
print("Portfolio Variance: {:,.9f}".format(portfolio_var))
print("Portfolio Volatility: {:,.9f}".format(portfolio_vol))
print("1-day VaR (99% confidence): ${:,.9f}".format(one_day_VaR))
```

```python
# historical VaR calculation
import pandas as pd
import numpy as np
from datetime import datetime

# Define the portfolio and historical dataframes
portfolio = {"JNJ": 1.2*10**6,
             "MRK": 2.5*10**6,
             "PFE": 3.5*10**6
            }
JNJ = pd.read_csv("hw3_JNJ.csv")
MRK = pd.read_csv("hw3_MRK.csv")
PFE = pd.read_csv("hw3_PFE.csv")

# Use the most recent market price as the current market price
current_prices = {"JNJ": JNJ.iloc[-1]["Adj Close"],
                  "MRK": MRK.iloc[-1]["Adj Close"],
                  "PFE": PFE.iloc[-1]["Adj Close"]
                 }

JNJ_preds = []
MRK_preds = []
PFE_preds = []

JNJ_rates = []
MRK_rates = []
PFE_rates = []

JNJ_pos_future = []
MRK_pos_future = []
PFE_pos_future = []

for i in range(-2,-len(JNJ),-1):
    JNJ_cur = JNJ.iloc[i]["Adj Close"]
    JNJ_prev = JNJ.iloc[i-1]["Adj Close"]
    JNJ_change = np.log(JNJ_cur) - np.log(JNJ_prev)
    JNJ_pred = current_prices["JNJ"]*np.exp(JNJ_change)
    JNJ_preds.append(JNJ_pred)
    JNJ_rates.append(JNJ_change)
    JNJ_pos_future.append((JNJ_pred-current_prices["JNJ"])*1200000)
    
    MRK_cur = MRK.iloc[i]["Adj Close"]
    MRK_prev = MRK.iloc[i-1]["Adj Close"]
    MRK_change = np.log(MRK_cur) - np.log(MRK_prev)
    MRK_pred = current_prices["MRK"]*np.exp(MRK_change)
    MRK_preds.append(MRK_pred)
    MRK_rates.append(MRK_change)
    MRK_pos_future.append((MRK_pred-current_prices["MRK"])*2500000)
    
    PFE_cur = PFE.iloc[i]["Adj Close"]
    PFE_prev = PFE.iloc[i-1]["Adj Close"]
    PFE_change = np.log(PFE_cur) - np.log(PFE_prev)
    PFE_pred = current_prices["PFE"]*np.exp(PFE_change)
    PFE_preds.append(PFE_pred)
    PFE_rates.append(PFE_change)
    PFE_pos_future.append((PFE_pred-current_prices["PFE"])*3500000)
    
portfolio_pos_future = np.array(JNJ_pos_future) + np.array(MRK_pos_future) + np.array(PFE_pos_future)

confidence_level = 0.99
one_day_VaR = np.percentile(portfolio_pos_future, 0.99)
one_day_VaR
```

(c) Reasoning

The two different VaR results that we obtained using the parametric VaR and the historical VaR are likely to be different due to the underlying assumptions and methodologies of the two approaches.

The parametric VaR approach relies on the normality assumption of the returns distribution and uses a covariance matrix to estimate the variance of the portfolio. This approach can be relatively quick and easy to compute, but its accuracy depends heavily on the normality assumption and the quality of the estimated covariance matrix.

On the other hand, the historical VaR approach is a non-parametric approach that does not require any assumptions about the underlying distribution of returns. Instead, it uses the historical returns data to estimate the loss distribution of the portfolio, and can provide a more accurate estimate of VaR in situations where the returns distribution is not normal or when there is a lack of data on the estimated covariance matrix.

In general, it is recommended to use a combination of different VaR approaches to gain a more complete understanding of the portfolio's risk profile. This can include stress-testing the portfolio using different scenarios, such as extreme market events, as well as using other risk metrics such as expected shortfall or conditional VaR.

Ultimately, the choice of which VaR approach to use will depend on the risk manager's specific needs and the characteristics of the portfolio being evaluated. It is important for the risk manager to carefully consider the assumptions and limitations of each method and to use multiple approaches to validate the results. 

But, I would advise the risk manager to consider using the parametric approach instead of the historical approach for calculating the 99% VaR. This is because the historical approach relies heavily on the limited data available and can produce unrealistic price changes if only a few exceptional cases occurred in the past. On the other hand, the parametric approach takes into account the properties of the entire dataset, rather than focusing on isolated extreme events, making it a more reliable prediction method.
